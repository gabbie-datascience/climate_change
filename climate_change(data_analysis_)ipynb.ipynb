## outliers analysis

```python
# imports
```

```python
import matplotlib.pyplot
import seaborn
```

```python
# columns
```

```python
categories=['Global_Warming_Temperature_Change_C', 'Extreme_Weather_Events_Count', 'Sea_Level_Rise_mm', 'Ocean_pH', 'Biodiversity_Loss_Species', 'Carbon_Emissions_Million_Tons', 'Heat_Related_Illnesses_Cases', 'Food_Water_Scarcity_Index', 'Displacement_Millions']
```

```python
# outliers analysis( with charts - using for )
```

```python
fig, axes=matplotlib.pyplot.subplots(3,3, figsize=(12,8))
axes=axes.flatten()

for i, column in enumerate(categories):
  axes[i].boxplot(climate_change_copy[column])
  axes[i].set_title(column)
  axes[i].set_ylabel('Values')

for j in range(len(categories), len(axes)):
  fig.delaxes(axes[j])

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

## line chart - temperature trends

```python
matplotlib.pyplot.figure(figsize=(10,6))
seaborn.lineplot(data=climate_change_copy, x='Year', y='Global_Warming_Temperature_Change_C')

matplotlib.pyplot.title('Temperature Trends')
matplotlib.pyplot.xlabel('Year')
matplotlib.pyplot.ylabel('Temperature Change (C)')

matplotlib.pyplot.show()
```

## line chart, bar chart - carbon emissions

```python
matplotlib.pyplot.figure(figsize=(10,6))
seaborn.lineplot(data=climate_change_copy, x='Year', y='Carbon_Emissions_Million_Tons', color='tab:red', marker='o')

matplotlib.pyplot.title('Carbon Emissions Over Time')
matplotlib.pyplot.xlabel('Year')
matplotlib.pyplot.ylabel('Carbon Emissions(Million Tons)')

matplotlib.pyplot.show()
```

```python
# bar chart - interval division
```

```python
intervals=[(1940, 1960), (1960, 1980), (1980, 2000), (2000, 2010), (2010, 2025)]
```

```python
# data separate
```

```python
climate_change_carbon_emissions=climate_change_copy[['Year', 'Carbon_Emissions_Million_Tons']]
climate_change_carbon_emissions
```

```python
climate_change_carbon_emissions['Year'].dtypes
```

```python
climate_change_carbon_emissions['Year']=pandas.to_datetime(climate_change_carbon_emissions['Year'])
climate_change_carbon_emissions.dtypes
```

```python
for start, end in intervals:
    start_date=pandas.to_datetime(start, format='%Y')
    end_date=pandas.to_datetime(end, format='%Y')

    climate_change_copy_subset = climate_change_carbon_emissions[
        (climate_change_carbon_emissions['Year'] >= start_date) &
        (climate_change_carbon_emissions['Year'] < end_date)
    ]

    matplotlib.pyplot.figure(figsize=(10, 6))
    seaborn.barplot(y='Year', x='Carbon_Emissions_Million_Tons', data=climate_change_copy_subset, color='tab:blue')

    matplotlib.pyplot.title(f'Carbon Emissions from {start} to {end}')
    matplotlib.pyplot.xlabel('Carbon Emissions (Million Tons)')
    matplotlib.pyplot.ylabel('Year')
    matplotlib.pyplot.show()
```
